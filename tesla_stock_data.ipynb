```python
import yfinance as yf

# Specify the ticker symbol and timeframe
ticker = "TSLA"
start_date = "2022-01-01"
end_date = "2023-01-01"

# Download historical stock data
tesla_data = yf.download(ticker, start=start_date, end=end_date)

# Reset the index
tesla_data.reset_index(inplace=True)

# Save the dataframe
tesla_data.to_csv("tesla_stock_data.csv", index=False)

# Display the first five rows
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://finance.yahoo.com/quote/TSLA/'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find the element containing revenue data (modify this based on the website structure)
revenue_element = soup.find('span', {'data-reactid': '50'})

# Check if the element is found
if revenue_element:
    # Extract the revenue value
    tesla_revenue = revenue_element.text
    print("Tesla Revenue:", tesla_revenue)
else:
    print("Revenue data not found on the page. Check the HTML structure.")
```
